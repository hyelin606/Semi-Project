```python
import pandas as pd

# ì£¼ì–´ì§„ ë°ì´í„°í”„ë ˆì„
df = pd.read_csv('data/new_dataframe.csv')

# ì‹œê°„ëŒ€ë¥¼ ë ˆì´ë¸” ì¸ì½”ë”©í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def encode_time(time_str):
    """
    ì‹œê°„ëŒ€ë¥¼ ë ˆì´ë¸” ì¸ì½”ë”©í•˜ëŠ” í•¨ìˆ˜.

    Parameters:
    - time_str (str): ì‹œê°„ëŒ€ ë¬¸ìì—´ (ì˜ˆ: '00:00')

    Returns:
    - int: ë ˆì´ë¸” ì¸ì½”ë”©ëœ ì‹œê°„ëŒ€ (1~6)
    """
    if '00' <= time_str < '06':
        return 1
    elif '06' <= time_str < '11':
        return 2
    elif '11' <= time_str < '14':
        return 3
    elif '14' <= time_str < '17':
        return 4
    elif '17' <= time_str < '21':
        return 5
    elif '21' <= time_str <= '24':
        return 6

# ìƒê¶Œ êµ¬ë¶„ ì½”ë“œë¥¼ ë ˆì´ë¸” ì¸ì½”ë”©í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def encode_market_code(code):
    """
    ìƒê¶Œ êµ¬ë¶„ ì½”ë“œë¥¼ ë ˆì´ë¸” ì¸ì½”ë”©í•˜ëŠ” í•¨ìˆ˜.

    Parameters:
    - code (str): ìƒê¶Œ êµ¬ë¶„ ì½”ë“œ (A, D, R, U)

    Returns:
    - int: ë ˆì´ë¸” ì¸ì½”ë”©ëœ ìƒê¶Œ êµ¬ë¶„ ì½”ë“œ (1~4)
    """
    if code == 'A':
        return 1
    elif code == 'D':
        return 2
    elif code == 'R':
        return 3
    elif code == 'U':
        return 4

# ì‹œê°„ëŒ€ë¥¼ ë ˆì´ë¸” ì¸ì½”ë”©í•˜ì—¬ ìƒˆë¡œìš´ ì»¬ëŸ¼ ìƒì„±
df['ì‹œê°„ëŒ€_ë ˆì´ë¸”'] = df['ì‹œê°„ëŒ€'].apply(encode_time)

# ìƒê¶Œ êµ¬ë¶„ ì½”ë“œë¥¼ ë ˆì´ë¸” ì¸ì½”ë”©í•˜ì—¬ ìƒˆë¡œìš´ ì»¬ëŸ¼ ìƒì„±
df['ìƒê¶Œ_êµ¬ë¶„_ì½”ë“œ_ë ˆì´ë¸”'] = df['ìƒê¶Œ_êµ¬ë¶„_ì½”ë“œ'].apply(encode_market_code)

# ê²°ê³¼ í™•ì¸
print(df[['ì‹œê°„ëŒ€', 'ì‹œê°„ëŒ€_ë ˆì´ë¸”']].head())
print(df[['ìƒê¶Œ_êµ¬ë¶„_ì½”ë“œ', 'ìƒê¶Œ_êµ¬ë¶„_ì½”ë“œ_ë ˆì´ë¸”']].head())

# ì‹œê°„ëŒ€ì™€ ìƒê¶Œ_êµ¬ë¶„_ì½”ë“œ ì»¬ëŸ¼ ì‚­ì œ
df.drop(['ì‹œê°„ëŒ€', 'ìƒê¶Œ_êµ¬ë¶„_ì½”ë“œ'], axis=1, inplace=True)

# íŒŒì¼ë¡œ ì €ì¥
df.to_csv('data/encoded_dataframe.csv', index=False)
```

```python
import warnings
import re
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from scipy.stats import boxcox
from scipy.special import inv_boxcox
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from lightgbm import LGBMRegressor
from sklearn.model_selection import GridSearchCV

# ê²½ê³  ë¬´ì‹œ
warnings.filterwarnings("ignore")

# í°íŠ¸ ì„¤ì •
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')

data = pd.read_csv('data/final_merged_update_store_age_df.csv')

# ìƒˆë¡œìš´ í”¼ì²˜ ì¶”ê°€
data['í¸ì˜ì _ë°€ë„'] = data['ìœ ì‚¬_ì—…ì¢…_ì í¬_ìˆ˜']/data['ì˜ì—­_ë©´ì ']

# ë¶ˆí•„ìš”í•œ ì»¬ëŸ¼ ì œê±°
data = data.drop(columns=['ì˜ì—­_ë©´ì ', 'ìƒê¶Œ_ì½”ë“œ', 'ìì¹˜êµ¬_ì½”ë“œ', 'í–‰ì •ë™_ì½”ë“œ', 'ìì¹˜êµ¬_ì½”ë“œ_ëª…', 'í‰ì¼_ìœ ë™ì¸êµ¬_ìˆ˜_í‰ê· ', 'ì£¼ë§_ìœ ë™ì¸êµ¬_ìˆ˜_í‰ê· ', 'ì†Œë“_êµ¬ê°„_ì½”ë“œ', 'ì í¬_ìˆ˜', 'ê°œì—…_ìœ¨', 'íì—…_ë¥ ', 'í”„ëœì°¨ì´ì¦ˆ_ì í¬_ìˆ˜'])

# ë²”ì£¼í˜• ë³€ìˆ˜ì™€ ìˆ«ìí˜• ë³€ìˆ˜ êµ¬ë¶„
cat_cols = ['ê¸°ì¤€_ë…„', 'ê¸°ì¤€_ë¶„ê¸°', 'ì‹œê°„ëŒ€', 'ìƒê¶Œ_êµ¬ë¶„_ì½”ë“œ_ëª…', 'ìƒê¶Œ_ì½”ë“œ_ëª…', 'í–‰ì •ë™_ì½”ë“œ_ëª…']
num_cols = data.columns.difference(cat_cols).tolist()


# ë…ë¦½ë³€ìˆ˜ì™€ ì¢…ì†ë³€ìˆ˜ ë¶„ë¦¬
y = data.pop('ì‹œê°„ëŒ€_ë§¤ì¶œê¸ˆì•¡')
X = data

# ë…ë¦½ë³€ìˆ˜ ë°ì´í„° ìŠ¤ì¼€ì¼ë§
# ë³€ìˆ˜ êµ¬ë¶„ ì½”ë“œì—ì„œ ì¢…ì†ë³€ìˆ˜ ì œê±°
num_cols.remove('ì‹œê°„ëŒ€_ë§¤ì¶œê¸ˆì•¡')

## ë²”ì£¼í˜• ë³€ìˆ˜ ë”ë¯¸í™”
X = pd.get_dummies(X, columns=cat_cols)

## ë”ë¯¸ ë³€ìˆ˜í™”ëœ ê°’ì´ ë¶ˆë¦¬ì–¸ í˜•íƒœë¡œ ë‚˜ì™”ë‹¤ë©´ 0ê³¼ 1ë¡œ ë³€í™˜
X.replace({True: 1, False: 0}, inplace=True)

## ìˆ«ìí˜• ë³€ìˆ˜ ì •ê·œí™”
scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])


# í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ë¶„í• 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)

def clean_feature_names(X):
    # íŠ¹ìˆ˜ ë¬¸ìë¥¼ ì œê±°í•˜ê³  ì–¸ë”ìŠ¤ì½”ì–´(_)ë¡œ ëŒ€ì²´
    X.columns = [re.sub(r'\W+', '_', col) for col in X.columns]
    return X

# ë°ì´í„°í”„ë ˆì„ Xì˜ íŠ¹ì„± ì´ë¦„ì„ ì •ì œ
X_train = clean_feature_names(X_train)
X_test = clean_feature_names(X_test)

# Box-Cox ë³€í™˜ ì ìš©
y_train_boxcox, lambda_ = boxcox(y_train + 1)  # +1ì„ ë”í•˜ëŠ” ì´ìœ ëŠ” ìŒìˆ˜ ê°’ì´ ì—†ë„ë¡ í•˜ê¸° ìœ„í•¨
y_test_boxcox = boxcox(y_test + 1, lmbda=lambda_)

# ëª¨ë¸ í•™ìŠµ - LightGBM
lgbm_regression = LGBMRegressor(random_state=42)

# íƒìƒ‰í•  í•˜ì´í¼íŒŒë¼ë¯¸í„° ë²”ìœ„ ì„¤ì •
param_grid = {
    'num_leaves': [25, 30, 35],
    'learning_rate': [0.12, 0.13, 0.14],
    'n_estimators': [375, 400, 425]
}

# GridSearchCV ê°ì²´ ìƒì„±
grid_search = GridSearchCV(estimator=lgbm_regression, param_grid=param_grid, cv=3)

# ëª¨ë¸ í•™ìŠµ
grid_search.fit(X_train, y_train_boxcox)

# ìµœì ì˜ ëª¨ë¸ë¡œ ì˜ˆì¸¡ ìˆ˜í–‰
best_lgbm_regression = grid_search.best_estimator_
y_pred_lgbm_grid_boxcox = best_lgbm_regression.predict(X_test)


# ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ì›ë˜ì˜ ìŠ¤ì¼€ì¼ë¡œ ë˜ëŒë¦¬ê¸° ìœ„í•´ ì—­ Box-Cox ë³€í™˜ ì ìš©
y_pred_lgbm_grid = inv_boxcox(y_pred_lgbm_grid_boxcox, lambda_)

# í‰ê°€
def evaluate(y_true, y_pred):
    # MSE ê³„ì‚°
    mse = mean_squared_error(y_true, y_pred)

    # MAE ê³„ì‚°
    mae = mean_absolute_error(y_true, y_pred)

    # R-squared ê³„ì‚°
    r_squared = r2_score(y_true, y_pred)

    # RMSE ê³„ì‚°
    rmse = np.sqrt(mse)

    return {
        'MSE': mse,
        'MAE': mae,
        'R-squared': r_squared,
        'RMSE': rmse
    }


evaluate(y_test, y_pred_lgbm_grid)
```
